```python
vendas_correto.describe()
```

```python
#vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado =='SP'").describe()
```

```python
#Departamento com mais vendas (unitarias)
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#Média de preço com frete por departamento
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
#quantidade de vendas por dia

vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
#quantidade de vendas por mês

# 1 criar coluna com o mês
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
# 2 agrupo pelo mês
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel("/content/drive/MyDrive/EstudosColab/cliente_varejo.xlsx")
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
# Transformar de INT para FLOAT
cliente = cliente.astype({"renda":"float"})
```

```python
cliente.head()
```

```python
vendas_correto.head()
```

### JOIN NO PYTHON

```python
vendas_cliente = vendas_correto.merge(cliente, how="left", on="cliente_Log")
```

```python
vendas_cliente.head()
```

```python
# media de renda por canal de venda
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index(),2)
agg_idcanal_renda.head()
```

```python
#Media de idade por bandeira

agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").reset_index(),2)
agg_idade_bandeira.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")
plt.ylabel("Média de Idade")
plt.title("Idade Média por Bandeira", loc="center", fontsize=18, color="black")
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="pink")
plt.ylabel("Renda Média")
plt.title("Renda Média por Canal de Venda", loc="center", fontsize=18, color="blue")
```

```python
# Configurar o tamanho da figura
plt.figure(figsize=(7, 4))

# Criar o gráfico de barras
bars = plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="pink")

# Adicionar os valores acima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color="black", fontsize=10)

# Configurar rótulos e título
plt.ylabel("Renda Média")
plt.title("Renda Média por Canal de Venda", loc="center", fontsize=18, color="blue")

# Exibir o gráfico
plt.show()
```
